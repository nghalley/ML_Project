```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

pd.options.display.max_columns = 200
```

```python
df = pd.read_csv(r'C:\Users\nobin\Desktop\file1\Affordable_Housing.csv')
df.fillna(0,inplace=True)
df.head(500)
```

```python
test = df[["AVG_Borrower_Gross_Income","AVG_Request_Amount", "State", "AVG_No_of_Dependent"]]

test["AVG_Borrower_Gross_Income"] = test["AVG_Borrower_Gross_Income"].str.replace(",","")
test["AVG_Borrower_Gross_Income"] = test.loc[:, "AVG_Borrower_Gross_Income"].astype(float)

test["AVG_Request_Amount"] = test["AVG_Request_Amount"].str.replace(",","")
test["AVG_Request_Amount"] = test.loc[:, "AVG_Request_Amount"].astype(float)

# test["AVG_No_of_Dependent"] = test["AVG_No_of_Dependent"].str.replace(",","")
# test["AVG_No_of_Dependent"] = test.loc[:, "AVG_No_of_Dependent"].astype(float)
```

```python
import matplotlib
cmap = matplotlib.colors.Colormap('jet', N=test.State.nunique())
```

```python
for i in test.State.unique():
    plt.scatter(test[test.State == i].AVG_Borrower_Gross_Income, test[test.State == i].AVG_Request_Amount, cmap=cmap);
plt.title("correlation between AVG_Borrower_Gross_Income and AVG_Request_Amount")
plt.xlabel("AVG_Borrower_Gross_Income")
plt.ylabel("AVG_Request_Amount")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.model_selection import train_test_split
```

```python
test = test.dropna()

X = test[['AVG_Borrower_Gross_Income', 'AVG_No_of_Dependent']].values
Y = test.AVG_Request_Amount.values.reshape(-1,1)

X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.30, random_state=0)
```

```python
clf = LR()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor as RNR
rf = RNR(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
